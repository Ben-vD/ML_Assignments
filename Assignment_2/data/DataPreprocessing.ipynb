```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, OrdinalEncoder

# Read the Excel file  # Replace with your file path
data = pd.read_excel("Dry_Bean_Dataset.xlsx")

# Check the first few rows of the dataset
print("Original Data:")
print(data.head())

# Split features and target variable
X = data.iloc[:, :-1]  # All rows, all columns except the last one (features)
y = data.iloc[:, -1]   # All rows, only the last column (target)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Encode the target variable (assuming it is categorical)
encoder = OrdinalEncoder()
y_encoded = encoder.fit_transform(y.values.reshape(-1, 1))

# Create a new DataFrame with scaled features and encoded target
output_data = pd.DataFrame(X_scaled, columns=X.columns)
output_data['Target'] = y_encoded

# Write the output DataFrame to a new CSV file
output_file_path = 'standardized_output.csv'  # Specify output file name
output_data.to_csv(output_file_path, index=False, header=False)

print("Data has been standardized and saved to:", output_file_path)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Read in the dataset without headers
data = pd.read_csv('abalone.data', header=None)  # Update with the correct path to your dataset

# Separate features and target variable using indices
X = data.iloc[:, :-1]  # All rows, all columns except the last
y = data.iloc[:, -1]   # All rows, only the last column (target variable)

# Adjust the 'Rings' column to start at 0
min_ring_value = y.min()  # Find the minimum value in the 'Rings' column
y = y - min_ring_value  # Subtract the minimum value from each entry

# Print the number of unique classes in the target variable 'Rings'
unique_classes_y = y.unique()  # Get unique values in the adjusted 'Rings' column
print(f"Unique classes in 'Rings': {unique_classes_y}")
print(f"Number of unique classes: {len(unique_classes_y)}")

# Print the number of features
num_features = X.shape[1]  # Number of columns in X
print(f"Number of features: {num_features}")

# Encode the 'Sex' field
label_encoder = LabelEncoder()
X.iloc[:, 0] = label_encoder.fit_transform(X.iloc[:, 0])  # Encode the 'Sex' column (assumed to be the first column)

# Normalize the feature columns using StandardScaler
scaler = StandardScaler()
X_standardized = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Combine standardized features with the target variable
standardized_data = pd.concat([X_standardized, y], axis=1)

# Store the standardized dataset in a CSV file
standardized_data.to_csv('abaloneNorm.csv', index=False)  # Saves the file without the index

# Display a confirmation message
print("Standardized dataset with encoded 'Sex' saved as 'abaloneNorm.csv'.")
```
