```python
from sklearn.metrics import r2_score

# Update 3 EquivDiameter outliers with regression. Correlated with Area and ConvexArea
# EquivDiameter = 2 * sqrt(Area / pi) 

from scipy.optimize import curve_fit

def AreaToEquivDiameter(area):
    return 2 * np.sqrt(area / np.pi)

x = df[(df["EquivDiameter"] < 600) & (df["EquivDiameter"] > 100)]["Area"]
y = df[(df["EquivDiameter"] < 600) & (df["EquivDiameter"] > 100)]["EquivDiameter"]

line = np.linspace(np.min(x), np.max(x), 10000)

plt.scatter(x, y)
plt.plot(line, AreaToEquivDiameter(line), color = "red")
plt.xlabel("Area")
plt.ylabel("EquivDiameter")
#plt.savefig("AreaEquivDiameterPlot.pdf", format="pdf", bbox_inches="tight")
plt.show()

imp = AreaToEquivDiameter(np.array([45750, 71368, 20464]))
print(imp)

df_clean.loc[[2762, 4780, 5884], "EquivDiameter"] = imp
df_clean.loc[[2762, 4780, 5884]]
```

```python
x = df_clean.loc[df_clean["ConvexArea"] != -30.0 , "Area"]
y = df_clean.loc[df_clean["ConvexArea"] != -30.0 , "ConvexArea"]
t = df_clean.loc[df_clean["ConvexArea"] == -30.0 , "Area"]

model = np.poly1d(np.polyfit(x, y, 1))
print(model)
polyline = np.linspace(np.min(x), np.max(x), 10000)
plt.scatter(x, y)
plt.plot(polyline, model(polyline), color='red')
plt.xlabel("Area")
plt.ylabel("ConvexArea")
print(r2_score(y, model(x)))

print(model(t))
#plt.savefig("AreaConvexAreaPlot.pdf", format="pdf", bbox_inches="tight")
plt.show()

df_clean.loc[df_clean["ConvexArea"] == -30.0 , "ConvexArea"] = int(model(t)[0])
df_clean.loc[df_clean["ConvexArea"] == -30.0 , "ConvexArea"]
```

```python
# Impute 18 Compactness missing values from ShapeFactor3 (no misssing)

def plotRegression(x, y, p, t):
    model = np.poly1d(np.polyfit(x, y, p))
    polyline = np.linspace(np.min(x), np.max(x), 10000)
    plt.scatter(x, y)
    plt.plot(polyline, model(polyline), color='red')
    plt.xlabel("ShapeFactor3")
    plt.ylabel("Compactness")
    #plt.savefig("CompactnessShapeFactor3.pdf", format="pdf", bbox_inches="tight")
    plt.show()
    print(r2_score(y, model(x)))
    print(model)
    return model(t)

x = df_clean.loc[~df["Compactness"].isna(), "ShapeFactor3"]
y = df_clean.loc[~df["Compactness"].isna(), "Compactness"]
t = df_clean.loc[df["Compactness"].isna(), "ShapeFactor3"]

imp = plotRegression(x, y, 2, t)

print(imp)

df_clean.loc[df_clean["Compactness"].isna(), "Compactness"] = imp
df_clean.loc[[104, 484, 788, 902, 1244, 2042, 2803, 3982, 5963, 8813, 9701, 9702, 9914, 10015, 10134, 10234, 10294, 10325]]
```

```python
def min_max_norm(data_col):
    col_min = np.min(data_col)
    col_max = np.max(data_col)
    return (data_col - col_min) / (col_max - col_min)

df_clean_OHE_MMN = df_clean_OHE.copy()
df_clean_OHE_MMN.iloc[:, np.r_[0:7, 8:19]] = df_clean_OHE_MMN.iloc[:, np.r_[0:7, 8:19]].apply(min_max_norm)
df_clean_OHE_MMN.shape
```
